```
"""
Сведения о компьютере:
OS:              Ubuntu 18.04.5 LTS
model name:      Intel(R) Xeon(R) CPU @ 2.30GHz
cpu MHz:         2299.998
cache size:      46080 KB
cpu cores:       1
cache_alignment: 64
MemTotal:        13333608 kB
MemFree:         10464560 kB
MemAvailable:    12411308 kB
"""

import time
import numpy as np
from numba import jit
import numba as nb
import string

m = 300

A = np.random.rand(m,m)
B = np.random.rand(m,m)

def f1_a(A,B):
    C = [[0 for row in range(len(A))] for col in range(len(B[0]))]
    for i in range(len(A)):
        for j in range(len(B[0])):
            tmp = 0.
            for k in range(len(B)):
                tmp += A[i][k]*B[k][j]
            C[i][j] = tmp

    return C

@jit
def f1_b(A,B):
    C = [[0 for row in range(len(A))] for col in range(len(B[0]))]
    for i in range(len(A)):
        for j in range(len(B[0])):
            tmp = 0.
            for k in range(len(B)):
                tmp += A[i][k]*B[k][j]
            C[i][j] = tmp
    return C

@jit(nopython = True)
def f1_c(A,B):
    C = [[0 for row in range(len(A))] for col in range(len(B[0]))]
    for i in range(len(A)):
        for j in range(len(B[0])):
            tmp = 0.
            for k in range(len(B)):
                tmp += A[i][k]*B[k][j]
            C[i][j] = tmp
    return C
"""
@jit('f8[:](f8[:],f8[:])')
def f1_d(A,B):
    C = [[0 for row in range(m-1)] for col in range(m-1)]
    for i in range(m-1):
        for j in range(m-1):
            tmp = 0.
            for k in range(m-1):
                tmp += A[i][k]*B[k][j]
            C[i][j] = tmp
    return C
"""
"""
Выдает ошибку:
<ipython-input-3-36599620d7e4>:58: NumbaWarning: 
Compilation is falling back to object mode WITH looplifting enabled because Function "f1_d" failed type inference due to: No implementation of function Function(<built-in function getitem>) found for signature:
 
 >>> getitem(float64, int64)
 
There are 22 candidate implementations:
      - Of which 22 did not match due to:
      Overload of function 'getitem': File: <numerous>: Line N/A.
        With argument(s): '(float64, int64)':
       No match.
"""

start_cpu = time.perf_counter()
start_real= time.time()

C = f1_a(A,B)
print(C)
t1_cpu = time.perf_counter()-start_cpu
t1_real = time.time()-start_real

L = open("numpy_test1w.txt", "w")

print("Название функции          | Действительное время в сек | Процессорное время в сек|След матрицы", file=L)
print("--------------------------|----------------------------|-------------------------|------------", file=L)
print("f1_a                      |{:8.6f}                   |{:8.6f}                |{:8.6f}".format(t1_real,t1_cpu,np.trace(C)), file=L)

start_cpu = time.perf_counter()
start_real= time.time()

C1 = f1_b(A,B)


t2_cpu = time.perf_counter()-start_cpu
t2_real = time.time()-start_real
print("f1_b @jit                 |{:8.6f}                    |{:8.6f}                 |{:8.6f}".format(t2_real,t2_cpu,np.trace(C1)), file=L)

t1_cpu = time.perf_counter()-start_cpu
t1_real = time.time()-start_real

start_cpu = time.perf_counter()
start_real= time.time()

C1 = f1_c(A,B)

t2_cpu = time.perf_counter()-start_cpu
t2_real = time.time()-start_real
print("f1_c @jit(nopython = True)|{:8.6f}                    |{:8.6f}                 |{:8.6f}".format(t2_real,t2_cpu,np.trace(C1)), file=L)

start_cpu = time.perf_counter()
start_real= time.time()

#C1 = f1_d(A,B)

t2_cpu = time.perf_counter()-start_cpu
t2_real = time.time()-start_real
print("f1_D @jit(nopython = True)|{:8.6f}                    |{:8.6f}                 |{:8.6f}".format(t2_real,t2_cpu,np.trace(C1)), file=L)

A = np.asarray(A)
B = np.asarray(B)

print("f2_a                      |{:8.6f}                    |{:8.6f}                 |{:8.6f}".format(t2_real,t2_cpu,np.trace(C1)), file=L)

start_cpu = time.perf_counter()
start_real= time.time()

C1 = f1_b(A,B)

t2_cpu = time.perf_counter()-start_cpu
t2_real = time.time()-start_real
print("f2_b @jit                 |{:8.6f}                    |{:8.6f}                 |{:8.6f}".format(t2_real,t2_cpu,np.trace(C1)), file=L)

t1_cpu = time.perf_counter()-start_cpu
t1_real = time.time()-start_real

start_cpu = time.perf_counter()
start_real= time.time()

C1 = f1_c(A,B)
print(C1)
t2_cpu = time.perf_counter()-start_cpu
t2_real = time.time()-start_real
print("f2_c @jit(nopython = True)|{:8.6f}                    |{:8.6f}                 |{:8.6f}".format(t2_real,t2_cpu,np.trace(C1)), file=L)


L.close()
```
